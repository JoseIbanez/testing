```python
%%sql 

SELECT eventTime,metric_value,source_id FROM nyc.meraki_31_latency LIMIT 10;

```

```python
%%sql
    
select 
    source_id,
    count(*) as hits, 
    avg(metric_value) as avg, 
    percentile(metric_value,.1) as p01, 
    percentile(metric_value,.5) as p05, 
    percentile(metric_value,.9) as p09,
    stddev(metric_value) as stddev
from nyc.meraki_31_latency 
where metric_value > 0
group by source_id
having hits > 100
order by avg
limit 500;
```
